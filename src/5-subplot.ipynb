### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import matplotlib.pyplot as plt

# Data for each plot
# 1. Line chart
x_line = [0, 1, 2, 3, 4, 5]
y_line = [0, 1, 4, 9, 16, 25]

# 2. Bar chart
categories = ['A', 'B', 'C', 'D']
bar_values = [5, 7, 3, 9]
bar_colors = ['red', 'blue', 'green', 'orange']

# 3. Scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# 4. Pie chart
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot: Line chart
axs[0, 0].plot(x_line, y_line, marker='o', color='purple')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].grid(True)

# Second subplot: Bar chart
axs[0, 1].bar(categories, bar_values, color=bar_colors)
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Category')
axs[0, 1].set_ylabel('Value')

# Third subplot: Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter, color='teal')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
# Annotate each scatter point
for xi, yi in zip(x_scatter, y_scatter):
    axs[1, 0].annotate(f'({xi},{yi})', (xi, yi), textcoords="offset points", xytext=(5,5), ha='left')

# Fourth subplot: Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()

```
